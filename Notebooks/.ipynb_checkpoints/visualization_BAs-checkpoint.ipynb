```python

# import libraries
import xarray as xr
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import geopandas as gpd

#for the gif
import os
import glob
import PIL
from PIL import Image, ImageEnhance, ImageOps
import imageio
```

```python
HomeDir= Path('/home/gerhardt/Study/MSc')
DataDir = Path('/home/gerhardt/Study/MSc/Data/SAAQIS')
FigDir = Path(HomeDir/'Figures')
GisDir = Path(HomeDir/'Data/GIS/Data')
RadarDir = Path(HomeDir/'Data/Radar')
FireDir = Path(HomeDir/'Data/FIRE')
```

## open the Zarr file

```python
#insert the path you have saved the cube or go to tutorial Seasfire_datacube_tutorial.ipynb to see how to access it
ds=xr.open_dataset('/home/gerhardt/Radlab/seasfire.zarr')
```

```python
min_lon = 12.21
min_lat = -35.91
max_lon = 37.21
max_lat = -21.91

mask_lon = (ds.longitude >= min_lon) & (ds.longitude <= max_lon)
mask_lat = (ds.latitude >= min_lat) & (ds.latitude <= max_lat)
```

```python
cropped_ds = ds.where(mask_lon & mask_lat, drop=True)
```

## We are making a gif in a loop with the total burned area (gwis_ba variable) for each year

## Similarly you can do the same for each variable you want

```python
## in SeasFire Datacube v2.0 the burned area variables would have the water bodies masked with ERA-5 land sea mask 
# although while resampling and taking the sum nan values turn to 0. You can skip this by adding skipna=False in the sum but that there is a bug for that in the last year if data 2020
mask= ds['lsm'][:,:]
```

```python
mask
```

```python
gwis=cropped_ds['gwis_ba']
```

```python
gwis[1,:,:].plot()
```

```python
#resample to get yearly data 
gwis_all=gwis.groupby('time.year').sum()
```

```python
gwis_all= gwis_all.where(mask>0.5)
```

```python
#check how it looks
gwis_all[19,:,:].plot()
```

```python
stations = gpd.read_file(DataDir/'SAAQIS_Stations_Level1.shp')
stations = stations.set_crs('ESRI:102022')

radar = gpd.read_file(GisDir/'Radar.shp')
radar = gpd.GeoDataFrame(radar, geometry=gpd.points_from_xy(radar.xcoord, radar.ycoord))
radar = radar.to_crs('Africa_Albers_Equal_Area_Conic')
```

```python
#make sure all images have the same shape and size
for i in files:
    im = Image. open(i)
    print(im. info['dpi'])
    print(im)
```

```python
#probably it would not be the same so you have to resize, make background white and save 
for i in files:
    im = Image. open(i)
    #print(im)
    #this is a size I choose you can choose a different one as well
    img_resized = im.resize((2179,1579)) 
    im = ImageOps.expand(img_resized)
    #make background white so that the titles will appear 
    fill_color=(255,255,255,255)
    background = Image.new(im.mode[:-1], im.size, fill_color)
    background.paste(im, im.split()[-1]) # omit transparency
    im = background
    #you can increase the dpi as well(300 is a very good one)
    new_img = im.save(i, dpi=(100,100))
```
